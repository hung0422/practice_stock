# 用 SQL 的方式合併兩個 DataFrames

```python
df_city = pd.DataFrame({
    'state': ['密蘇里州', '亞利桑那州', '肯塔基州', '紐約州'],
    'city': ['堪薩斯城', '鳳凰城', '路易維爾', '紐約市']
})
df_city
```

```python
df_info = pd.DataFrame({
    'city': ['路易維爾', '堪薩斯城', '鳳凰城'],
    'population': [741096, 481420, 4039182],
    'feature': list('abc')})
df_info
```

```python
pd.merge(left=df_city,
         right=df_info, 
         how='left', # left : left outer join , right：right outer join , outer: full outer join , inner： inner join
         on='city', # 透過此欄位合併
         indicator=True # 新增一個欄位顯示結果中每一列的來源 (預設是False, 可以不顯示)
        )
```

```python
# for迴圈  itertuples
for row in df_city.itertuples():
    print(f'{row.city}是{row.state}裏頭的一個城市')
```
