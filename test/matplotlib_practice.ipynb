# 圓餅圖

```python
products = ['a','b','c','d']
revenue = [20,50,40,70]
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels=products)
plt.show()
```

```python
# 百分佔比 autopct
plt.pie(revenue, labels=products, autopct='%.2f%%')
plt.show()
```

```python
# 調整大小 radius
plt.pie(revenue, labels=products, autopct='%.2f%%', radius=1.5)
plt.show()
```

```python
# 更改顏色 colors
colors = ['yellow','pink','blue','grey']
plt.pie(revenue, labels=products, autopct='%.2f%%', radius=1.5, colors=colors)
plt.show()
```

```python
# 切出來 explode
explode = [0,0,0,0.1]
plt.pie(revenue, labels=products, autopct='%.2f%%', radius=1.5, colors=colors, explode=explode)
plt.show()
```

```python
# 陰影 shadow
plt.pie(revenue, labels=products, autopct='%.2f%%', radius=1.5, colors=colors, explode=explode, shadow=True)
plt.show()
```

```python
revenue2 = [30,100,20,70]
fig,axis = plt.subplots(1,2)
axis[0].pie(revenue, labels=products, autopct='%.2f%%', radius=1.5, colors=colors, explode=explode, shadow=True)
axis[0].set_title('title1')
axis[1].pie(revenue2, labels=products, autopct='%.2f%%', radius=1.2, colors=colors, explode=explode, shadow=True)
axis[1].set_title('title2')
plt.show()
```

# 雙環圖

```python
class_1=['Finance','Industry','IT']
class_2=['Banks','Securities','Cars','Oils','Computers','5G']

val_1=[80,70,100]
val_2=[30,50,35,35,30,70]
```

```python
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7)
plt.show()
```

```python
# 圓環圖
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'))
plt.show()
```

```python
# 雙環圖
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'))
ax.pie(val_2, labels=class_2, autopct = '%.2f%%', radius=1, wedgeprops=dict(width=0.3, edgecolor='w'))
plt.show()
```

```python
# 調整圖例和值的距離 labeldistance, pctdistance
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, labeldistance=0.65, pctdistance=0.35, wedgeprops=dict(width=0.3, edgecolor='w'))
ax.pie(val_2, labels=class_2, autopct = '%.2f%%', radius=1,pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor='w'))
plt.show()
```

```python
# 調整圖例或值的大小
fig,ax = plt.subplots(figsize=(5,5))
pie1,t1_text,n1_text=ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, labeldistance=0.65, pctdistance=0.35, wedgeprops=dict(width=0.3, edgecolor='w'))
ax.pie(val_2, labels=class_2, autopct = '%.2f%%', radius=1,pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor='w'))

for t in t1_text:
    t.set_size(13)
    
for n in n1_text:
    n.set_size(13)
    
plt.show()
```

```python
# title
fig,ax = plt.subplots(figsize=(5,5))
pie1,t1_text,n1_text=ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, labeldistance=0.65, pctdistance=0.35, wedgeprops=dict(width=0.3, edgecolor='w'))
ax.pie(val_2, labels=class_2, autopct = '%.2f%%', radius=1,pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor='w'))

for t in t1_text:
    t.set_size(13)
    
for n in n1_text:
    n.set_size(13)
    
ax.set(title='title', aspect='equal')
    
plt.show()
```

```python
# colors
colors=['lightblue','skyblue','lightcoral','orange','lightgreen','limegreen']
fig,ax = plt.subplots(figsize=(5,5))
pie1,t1_text,n1_text=ax.pie(val_1, labels=class_1, autopct = '%.2f%%', radius=0.7, labeldistance=0.65, pctdistance=0.35, wedgeprops=dict(width=0.3, edgecolor='w'))
ax.pie(val_2, labels=class_2, autopct = '%.2f%%', radius=1,pctdistance=0.9,colors=colors, wedgeprops=dict(width=0.3, edgecolor='w'))

for t in t1_text:
    t.set_size(13)
    
for n in n1_text:
    n.set_size(13)
    
ax.set(title='title', aspect='equal')
    
plt.show()
```

# 散佈圖

```python
return1=normal(size=100)*10
risk1=uniform(0,0.5,size=100)

return2=normal(0.5,size=100)*10
risk2=uniform(0.3,1,size=100)

return3=normal(-2,size=100)*10
risk3=uniform(0,1,size=100)
```

```python
plt.scatter(risk1, return1)
plt.xlabel('risk')
plt.ylabel('return')
plt.show()
```

```python
plt.scatter(risk1, return1)
plt.scatter(risk2, return2)
plt.scatter(risk3, return3)
plt.xlabel('risk')
plt.ylabel('return')
plt.show()
```

```python
plt.scatter(risk1, return1, label='stock1')
plt.scatter(risk2, return2, label='stock2')
plt.scatter(risk3, return3, label='stock3')
plt.xlabel('risk')
plt.ylabel('return')
plt.legend()
plt.show()
```

```python
plt.scatter(risk1, return1, label='stock1')
plt.scatter(risk2, return2, label='stock2')
plt.scatter(risk3, return3, label='stock3', color='red', marker='*', s=100)
plt.xlabel('risk', size=15)
plt.ylabel('return', size=15)
plt.title('test', size=30, color='limegreen')
plt.legend()
plt.show()
```

```python
# 透明度 alpha
plt.scatter(risk1, return1, label='stock1', alpha=0.5)
plt.scatter(risk2, return2, label='stock2')
plt.scatter(risk3, return3, label='stock3', color='red', marker='*', s=100)
plt.xlabel('risk', size=15)
plt.ylabel('return', size=15)
plt.title('test', size=30, color='limegreen')
plt.legend()
plt.show()
```

```python
# 坐標軸
plt.scatter(risk1, return1, label='stock1', alpha=0.5)
plt.scatter(risk2, return2, label='stock2')
plt.scatter(risk3, return3, label='stock3', color='red', marker='*', s=100)
plt.xlabel('risk', size=15)
plt.ylabel('return', size=15)
plt.title('test', size=30, color='limegreen')
plt.xlim(xmin=0.2, xmax=0.8)
plt.ylim(ymin=-60, ymax=20)
plt.legend()
plt.show()
```

```python
# 軸的大小
plt.scatter(risk1, return1, label='stock1', alpha=0.5)
plt.scatter(risk2, return2, label='stock2')
plt.scatter(risk3, return3, label='stock3', color='red', marker='*', s=100)
plt.xlabel('risk', size=15)
plt.ylabel('return', size=15)
plt.title('test', size=30, color='limegreen')
plt.xlim(xmin=0.2, xmax=0.8)
plt.ylim(ymin=-60, ymax=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
# 網格
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(data['Close'], color='skyblue', label='close')
ax.set_ylabel('close', color='skyblue', fontsize=20)
ax.tick_params(axis="y", labelcolor='skyblue')
plt.legend()
plt.grid()

ax2 = ax.twinx()
ax2.plot(data['Open'], color='pink', label='open')
ax2.set_ylabel('Open', color='pink', fontsize=20)
ax2.tick_params(axis="y", labelcolor='pink')
ax2.legend(loc='lower right')

plt.show()
```

# 直方圖

```python
data = pd.read_csv(r'./regression/lstm2.csv')
```

```python
data['Close'].hist()
plt.show()
```

```python
# 分成幾個群組 bins
data['Close'].hist(bins=20)
plt.show()
```

```python
# 分成幾個群組 (自訂) bins
bins = [9000,11000,12000,13000,14000,15000]
data['Close'].hist(bins=bins)
plt.show()
```

```python
# density 總面積加起來變成 1
data['Close'].hist(bins=20, density=True)
plt.show()
```

```python
# 顏色
data['Close'].hist(bins=20, density=True, color='skyblue')
plt.show()
```

```python
# range
data['Close'].hist(bins=20, density=True, color='skyblue', range=(13000,16000))
plt.show()
```

```python
# 累加圖 cumulative
data['Close'].hist(bins=20, density=True, color='skyblue', range=(13000,16000), cumulative=True)
plt.show()
```

```python
# 轉向 orientation
data['Close'].hist(bins=20, density=True, color='skyblue', range=(13000,16000), cumulative=True, orientation='horizontal')
plt.show()
```

```python
# 透明度 alpha
data['Close'].hist(bins=20, density=True, color='skyblue', alpha=0.5)
plt.show()
```
